```python
if isinstance(movies, pd.DataFrame) and isinstance(credits, pd.DataFrame):
    # Merge the DataFrames
    movies = movies.merge(credits, on='title')
    print(movies.shape)  # This will print the shape of the merged DataFrame
else:
    print("One of the variables is not a DataFrame. Please check the inputs.")
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast #converts string of list into list for which we use literal_eval function of ast module
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #split data into comma sperated list
movies.head()
```

```python
#remove the space between science fiction so that science and fiction do not treated as two different entity
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x)) #convert the list into string
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower()) #convert string into lower case
new_df.head(5)
```

vectorization using bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

MODEL BUILDING

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity #cosine similarity is different from cosine distance,in cosine similarity 1 means similarity is high 0 means similarity is low
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances=similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
recommend('Avatar')
```
